```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    'weather': ['sunny', 'sunny', 'sunny', 'rainy', 'rainy', 'rainy'],
    'activity': ['picnic', 'drink', 'errands', 'movie', 'movie', 'movie'],
    'decision': ['go out', 'go out', 'go out', 'stay in', 'stay in', 'stay in']
}

df = pd.DataFrame(data)

# Step 2: Convert categorical data to numerical
df['weather_code'] = df['weather'].map({'sunny': 0, 'rainy': 1})
df['decision_code'] = df['decision'].map({'go out': 1, 'stay in': 0})

# Features and target
X = df[['weather_code']]
y = df['decision_code']

# Step 3: Train the Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=42)
clf.fit(X, y)

# Step 4: Visualize the decision tree
plt.figure(figsize=(8,6))
plot_tree(clf, feature_names=['weather'], class_names=['stay in', 'go out'], filled=True, rounded=True, fontsize=12)
plt.title('Decision Tree for Activity Based on Weather')
plt.show()

# Step 5: Print tree rules
tree_rules = export_text(clf, feature_names=['weather'])
print(tree_rules)
```
