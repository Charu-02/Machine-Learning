```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Generate random data points
np.random.seed(42)
X = np.random.uniform(-3, 3, (200, 1))
y = (X[:, 0]**2 + np.random.randn(200) * 2 > 4).astype(int)  # Quadratic boundary with noise

# Train SVM with RBF kernel
model = SVC(kernel='rbf', C=1.0, gamma=1.0)
model.fit(X, y)

# Plot the data and decision boundary
xx = np.linspace(-3.5, 3.5, 500).reshape(-1, 1)
Z = model.predict(xx)

plt.scatter(X, y, c=y, cmap='bwr', edgecolors='k')
plt.plot(xx, Z, color='black')
plt.title("SVM Classifier on Quadratic Data")
plt.xlabel("Feature")
plt.ylabel("Class")
plt.show()
```
